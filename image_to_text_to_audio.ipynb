```python
import pytesseract
from PIL import Image
import gtts
from IPython.display import Audio
```

```python
img = Image.open('/content/Picture1.png')
print(img)
```

```python
result = pytesseract.image_to_string(img)
print(result)
```

```python
tts = gtts.gTTS(result)
tts.save("hello.mp3")
```

```python
Audio("hello.mp3", autoplay=True)
```
